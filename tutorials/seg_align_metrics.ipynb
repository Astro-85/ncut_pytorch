```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

import sys, os
sys.path.append(os.path.join(os.path.dirname(sys.path[0]), 'ncut_pytorch'))

from ncut_pytorch import NCUT, rgb_from_tsne_3d
from ncut_pytorch.backbone import list_models, load_model, extract_features

device = torch.device('cuda:1')
torch.cuda.set_device(device)
```

```python
list_models()
```

```python
model = load_model(model_name="SAM(sam_vit_b)")
images = torch.rand(20, 3, 1024, 1024)
model_features = extract_features(images, model, node_type='attn', layer=6, device=device)
# model_features = model(images)['attn'][6]  # this also works

inp = model_features.reshape(-1, 768)  # flatten
eigvectors, eigvalues = NCUT(num_eig=100, device=device).fit_transform(inp)
tsne_x3d, tsne_rgb = rgb_from_tsne_3d(eigvectors, device=device)

eigvectors = eigvectors.reshape(20, 64, 64, 100)  # (B, H, W, num_eig)
tsne_rgb = tsne_rgb.reshape(20, 64, 64, 3)  # (B, H, W, 3)
```
